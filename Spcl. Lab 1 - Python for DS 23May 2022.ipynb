```python
Create_Table = """ create table customer(CUS_ID int primary key , CUS_NAME varchar(50),CUS_GENDER varchar(20), 
                          CUS_CITY varchar(50), CUS_PHONE bigint);

create table category(CAT_ID int primary key, CAT_NAME varchar(50));

create table product(PRO_ID INT primary key, PRO_NAME varchar(50), PRO_DESC varchar(50), CAT_ID int, 
foreign Key(CAT_ID) references category(CAT_ID));

create table supplier(SUPP_ID int primary key, SUPP_NAME varchar(50), SUPP_CITY varchar(50), SUPP_PHONE bigint);

create table productdetails(PROD_ID int primary key, price float, PRO_ID int, SUPP_ID int, 
foreign key(PRO_ID) references product(PRO_ID), foreign Key(SUPP_ID) references supplier(SUPP_ID));

create table rating(RAT_ID int primary key, RAT_RATSTARS float, CUS_ID int, SUPP_ID int, 
foreign key(CUS_ID) references customer(CUS_ID), foreign key(SUPP_ID) references supplier(SUPP_ID));

create table orders(ORD_ID int primary key, ORD_AMOUNT float, ORD_DATE date, CUS_ID int, PROD_ID int,
foreign key(CUS_ID) references customer(CUS_ID), foreign key(PROD_ID) references productdetails(PROD_ID));

"""
for result in cursorObject.execute(Create_Table, multi=True):
    pass
```

```python
# Q3) Display the number of the customer group by their genders who 
#have placed any order of amount greater than or equal to Rs.3000.
connection = mysql.connector.connect(host="localhost",
                                     user="root",
                                     passwd="password",
                                     database="e_commerce")
cursorObject= connection.cursor()

question_three="""select count(cus_name), CUS_Gender from customer
join orders on customer.CUS_ID=orders.CUS_ID
where orders.ORD_AMOUNT>=3000
group by customer.CUS_gender"""

cursorObject.execute(question_three);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['NumberOfCustomers','Gender'])
output_df
```

```python
### Q4) Display all the order along with product name ordered by a customer having Customer_Id=2;

question_four="""select orders.ord_id as OrderID,Product.pro_name as ProductName,customer.cus_name as CustomerName
from customer
join orders on customer.CUS_ID=orders.cus_id
join productdetails on orders.prod_id = productdetails.PROD_ID
join product on productdetails.pro_id= product.pro_id
where customer.cus_id=2;"""

cursorObject.execute(question_four);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['OrderID','ProductName','CustomerName'])
output_df
```

```python
### Q5) Display the Supplier details who can supply more than one product.


question_five="""select supp_name supplierName, count(*) as TotalProducts from supplier
join productdetails on productdetails.SUPP_ID=supplier.SUPP_ID
group by productdetails.supp_id
having count(*)>1;
"""

cursorObject.execute(question_five);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['Supplier Name','Total Products'])
output_df
```

```python
### Q6) Find the category of the product whose order amount is minimum.

question_six="""select category.cat_name  from orders
join productdetails on orders.prod_id = productdetails.PROD_ID
join product on productdetails.pro_id= product.pro_id
join category on product.cat_id=category.CAT_ID
where orders.ord_amount = (select min(ord_amount) from orders);
"""

cursorObject.execute(question_six);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['Category Name'])
output_df
```

```python
### Q7) Display the Id and Name of the Product ordered after “2021-10-05”.

question_seven="""select product.Pro_name as ProductName, product.Pro_id as ProductID from orders
join productdetails on orders.prod_id = productdetails.PROD_ID
join product on productdetails.pro_id= product.pro_id
where orders.ord_date>'2021-10-05';
"""

cursorObject.execute(question_seven);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['Product Name','Product ID'])
output_df
```

```python
### Q8) Print the top 3 supplier name and id and rating on the basis of their rating along with the customer name who has given the rating.

question_eight="""select customer.cus_name as CustomerName, Supplier.SUPP_NAME as SupplierName, 
supplier.supp_id as SupplierID, rating.RAT_RATSTARS  from customer
join rating on rating.cus_id=customer.cus_id
join supplier on rating.supp_id=supplier.SUPP_ID
and rating.RAT_RATSTARS in (select rat_ratstars from (select rat_ratstars from rating  order by rat_ratstars desc limit 3 )as FindTop3Rat);"""

cursorObject.execute(question_eight);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['Customer Name','Supplier Name','Supplier ID','Rating'])
output_df
```

```python
## Q9) Display customer name and gender whose names start or end with character 'A'.


question_nine="""select cus_name as CustomerName, cus_gender as CustomerGender from customer
where cus_name regexp '^A|A$';"""

cursorObject.execute(question_nine);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['Customer Name','Customer Gender'])
output_df
```

```python
### Q10) Display the total order amount of the male customers.

question_ten="""select sum(ORD_AMOUNT) from customer
join orders on customer.CUS_ID=orders.CUS_ID
where customer.cus_gender='M';"""

cursorObject.execute(question_ten);
output= cursorObject.fetchall()

output_df = pd.DataFrame(output, columns=['Order Amount'])
output_df
```
